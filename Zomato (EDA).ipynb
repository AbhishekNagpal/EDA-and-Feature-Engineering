```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_val
```

```python
# plotting pie chart for top 5 countries

plt.pie(country_val[:5],labels=country_names[:5],autopct='%1.2f%%') # autopct is displaying numbers on pie chat
plt.show()
```

```python
# Observation - 
## Zomato have max orders from India then USA then UK and so on.
```

```python
final_df.columns
```

```python
# lets do something with ratings
```

```python
#.size() is counting the number of ratings
#.reset_index() is resetting the index
#.rename() is used to rename the column which is named as 0

ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
# Observation 
## maximum number of ratings between 2.5-3.4
# unrated count is very high
```

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
